```python
import requests
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
import selenium
from selenium import webdriver
from selenium.webdriver.chrome.service import Service as ChromeService
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
driver = webdriver.Chrome(service=ChromeService(ChromeDriverManager().install()))
driver.get('http://cricbuzz.com')
searchBox=driver.find_element(By.ID,'search_bar_menu')
inp=input()
searchBox.send_keys(inp)
searchBox.send_keys(Keys.ENTER)
```

```python
content=driver.page_source
soup=BeautifulSoup(content,'html.parser')
soup=soup.find_all("div",attrs={"class":"cb-search-list-item-wrap ng-scope"})
url=soup[0].find('a').get('href')
URL='http://cricbuzz.com'+url
#URL="https://www.cricbuzz.com/profiles/1413/virat-kohli"
headers={'User-Agent':'Chrome'}
response=requests.get(URL,headers=headers)
content=response.content
soup=BeautifulSoup(content,'html.parser')
page=soup.find("div",attrs={'class':'page'})
playerHeader=page.find("div",attrs={'class':'cb-col cb-col-100 cb-bg-white'})
playerDetails=page.find("div",attrs={'class':'cb-col cb-col-33 text-black'})
playerStats=page.find("div",attrs={'class':'cb-col cb-col-67 cb-bg-white cb-plyr-rt-col'})
detailsHeading=playerDetails.find_all("div",attrs={'class':'cb-col cb-col-40 text-bold cb-lst-itm-sm'})
detailsValues=playerDetails.find_all("div",attrs={'class':'cb-col cb-col-60 cb-lst-itm-sm'})
dataframe={}
dataframe['Name']=playerHeader.find("h1").text
for i,j in zip(detailsHeading,detailsValues):
    L=[]
    L.append(j.text)
    dataframe[i.text]=L
dataframe
personalDetails=pd.DataFrame.from_dict(dataframe)
personalDetails=personalDetails.T.rename(columns={0:' '})
print("\n\n==================================================Personal Details==================================================")
display(personalDetails)
statsTable=playerStats.find_all("table",attrs={'class':'table cb-col-100 cb-plyr-thead'})
battingColumns=[x.text for x in statsTable[0].find_all("th",attrs={'class','text-right'})]
formatDetails=[x.text for x in statsTable[0].find_all("td",attrs={'class','cb-col-8'})]
values=[x.text for x in statsTable[0].find_all("td",attrs={'class','text-right'})]
values=np.array(values).reshape([4,13])
data={i:list(j) for i,j in zip(formatDetails,values)}
df=pd.DataFrame.from_dict(data)
df['Format']=battingColumns
battingSummary=df.set_index('Format').T
print("\n\n==================================================Batting Summary==================================================")
display(battingSummary)
bowlingColumns=[x.text for x in statsTable[1].find_all("th",attrs={'class','text-right'})]
bowlingformatDetails=[x.text for x in statsTable[1].find_all("td",attrs={'class','cb-col-8'})]
bowlingvalues=[x.text for x in statsTable[1].find_all("td",attrs={'class','text-right'})]
bowlingvalues=np.array(bowlingvalues).reshape([4,12])
data={i:list(j) for i,j in zip(bowlingformatDetails,bowlingvalues)}
df=pd.DataFrame.from_dict(data)
df['Format']=bowlingColumns
bowlingSummary=df.set_index('Format').T
print("\n\n==================================================Bowling Summary==================================================")
display(bowlingSummary)
```
